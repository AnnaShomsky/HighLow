```python
    add_padding()
    print("The goal is to guess a random number. You want your ratio of guess to possible numbers to be as low as possible.\n You can get to pick your range.\n You will choose a number between 0 and a number of your choosing.")
    you_done=0
    List_of_ratios=[]
    import numpy as np
    import random
    while you_done<3:
        
        input_max= input("pick a maximum number.\n")
        input_max=int(input_max)
        x = random.randint(0, input_max)
    
        you_won = False
   

        #print(x)
        y = x

    # take input from a user
        print("input a number between 0 and " + str(input_max))

  

        list_of_inputs = []
        while (you_won == False):
            input_a = input()

            # type cast into integer
            input_a = int(input_a)
      
            list_of_inputs.append(input_a)
            if input_a == y:
                print("you win")
                you_won = True
            elif input_a < y:
                print(str(input_a)+ " is too low")
            else:
                print(str(input_a)+ " is too high")
       
    #plot your path
    
        from matplotlib import pyplot as plt
        
    
        plt.plot(list_of_inputs)
        plt.title("your path")
        plt.xlabel("number of guesses")
        plt.ylabel("number you guessed")
        plt.show()
    
    
    
    #show your inputs
        print("your guesses were: ")
        print(list_of_inputs)
        print("you found the number in " + str(len(list_of_inputs)) + " attempts" )
    
        length= len(list_of_inputs)
 #show your ratio
        ratio= round(int(length)/int(input_max),3)
        print("the ratio of guesses to possibilities is " + str(ratio))
        List_of_ratios.append(ratio)
    
        
   
    #iterate   
        you_done +=1
    
 #plot all your scores
plt.plot(List_of_ratios)
plt.title("your scores")
plt.xlabel("number of games")
plt.ylabel("your scores \n")
plt.show()  

avg= round(sum(List_of_ratios)/len(List_of_ratios),2)
#show best score
print("your best score is " + str(min(List_of_ratios)))
print("your worst score is " + str(max(List_of_ratios)))
print ("your average score is " + str(avg) + "\n");
 
```
